```python
def tiff_to_png_by_case(input_dir, output_dir):
    os.makedirs(output_dir, exist_ok=True)
    
    for file in os.listdir(input_dir):
        if not file.lower().endswith((".tif", ".tiff")):
            continue
        
        # Example filename: 001_01_L_01.tiff
        parts = file.split("_")
        
        # Extract case_id, sequence, and side: "001_01_L"
        case_id = parts[0]      # "001"
        sequence = parts[1]     # "01"
        side = parts[2]         # "L" or "R"
        image_num = parts[3].split(".")[0]  # "01" (remove extension)
        
        # Create folder name: "001_01_L"
        folder_name = f"{case_id}{sequence}{side}"
        case_folder = os.path.join(output_dir, folder_name)
        os.makedirs(case_folder, exist_ok=True)
        
        # Read TIFF and save as PNG with just the image number
        tiff_path = os.path.join(input_dir, file)
        png_name = f"{image_num}.png"
        png_path = os.path.join(case_folder, png_name)
        
        with Image.open(tiff_path) as img:
            img.save(png_path, format="PNG")
        
        print(f"Saved: {png_path}")

# Example usage
tiff_to_png_by_case(
    input_dir="VLM_Reasoning/data/raw",
    output_dir="VLM_Reasoning/data/processed"
)

```
